```python
X = np.array(data_mod[['Cases/population']])
where_are_NaNs = np.isnan(X)
X[where_are_NaNs] = 0

kmeans = KMeans(n_clusters=5, random_state=4).fit(X)
fig = plt.figure(figsize=[10, 6])
ax = fig.add_subplot(1, 1, 1)
ax.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap=cm.get_cmap('gist_ncar'));
```

```python
index_number = 0
for country in countries:
    temp_segment = data_time[data_time['Entity'] == country].copy()
    for i in range(len(temp_segment)-1):
        the_date = data_time.at[index_number,'Date']
        check_date = datetime.timedelta(1)
        next_date = data_time.at[index_number+1,'Date'] 
        if next_date != the_date + check_date:
            new_date = the_date + check_date 
            new_row = pd.DataFrame({'Entity':[country], 'Date':[new_date], 'Cases':[0]})
            print(new_row)
            data_time.append(new_row, ignore_index=True)   
        index_number += 1
    index_number += 1
    
    
```
